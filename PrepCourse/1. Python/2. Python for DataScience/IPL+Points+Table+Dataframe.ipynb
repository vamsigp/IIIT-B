```python
ipl17 = pd.DataFrame({'Team': ['MI', 'RPS', 'SRH', 'KKR', 'KXIP', 'DD', 'GL', 'RCB'],
                         'Matches': [14, 14, 14, 14, 14, 14, 14, 14],
                         'Won': [10, 9, 8, 8, 7, 6, 4, 3],
                         'Lost': [4, 5, 5, 6, 7, 8, 10, 10],
                         'Tied': [0, 0, 0, 0, 0, 0, 0, 0],
                         'N/R': [0, 0, 1, 0, 0, 0, 0, 1],
                         'Points': [20, 18, 17, 16, 14, 12, 8, 7],
                         'NRR': [0.784, 0.176, 0.469, 0.641, 0.123, -0.512, -0.412, -1.299],
                         'For': [2407, 2180, 2221, 2329, 2207, 2219, 2406, 1845],
                         'Against': [2242, 2165, 2118, 2300, 2229, 2255, 2472, 2033]},
                         index = range(1,9)
                    )

ipl17
```

```python
ipl = pd.merge(ipl17, ipl18, how='outer', on='Team')
ipl
```

```python
ipl.fillna(0, inplace=True)
ipl
```

```python
ipl['Matches'] = ipl['Matches_x']+ipl['Matches_y']
ipl.drop(labels=['Matches_x', 'Matches_y'], axis=1, inplace=True)
ipl
```

```python
ipl['won'] = ipl['Won_x']+ipl['Won_y']
ipl.drop(labels=['Won_x', 'Won_y'], axis=1, inplace=True)
ipl
```

```python
ipl.columns
```

```python
ipl['Points'] = ipl['Points_x']+ipl['Points_y']
ipl['NRR'] = ipl['NRR_x']+ipl['NRR_y']
ipl['For'] = ipl['For_x']+ipl['For_y']
ipl['Against'] = ipl['Against_x']+ipl['Against_y']
ipl.drop(labels=['Against_x', 'Against_y','Points_x','Points_y','NRR_x','NRR_y','For_x','For_y'], axis=1, inplace=True)
ipl
```

```python
ipl['N/R'] = ipl['N/R_x']+ipl['N/R_y']
ipl.drop(labels=['N/R_x', 'N/R_y'], axis=1, inplace=True)
ipl
```

```python
ipl['tied'] = ipl['Tied_x']+ipl['Tied_y']
ipl.drop(labels=['Tied_x', 'Tied_y'], axis=1, inplace=True)
ipl
```

```python
ipl['lost'] = ipl['Lost_x']+ipl['Lost_y']
ipl.drop(labels=['Lost_x', 'Lost_y'], axis=1, inplace=True)
ipl
```

```python
ipl
```

```python
ipl.sort_values(by=['Points'], ascending=False, inplace=True)
ipl
```

```python
ipl['winPercent'] = round(100*ipl['won']/ipl['Matches'],2)
ipl
```
