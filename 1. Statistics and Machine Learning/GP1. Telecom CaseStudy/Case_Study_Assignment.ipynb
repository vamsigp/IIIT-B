```python
telecom_data = telecom_data.drop('arpu_2g_9', axis=1)
telecom_data = telecom_data.drop('arpu_2g_8', axis=1)
# vamsi - night_pck_user_* null can be replaced by no/0
telecom_data = telecom_data.drop('night_pck_user_8', axis=1)
telecom_data = telecom_data.drop('night_pck_user_9', axis=1)

telecom_data = telecom_data.drop('arpu_3g_9', axis=1)
telecom_data = telecom_data.drop('arpu_3g_8', axis=1)
# vamsi - fb_user_* null can be replaced by no/0
telecom_data = telecom_data.drop('fb_user_8', axis=1)
telecom_data = telecom_data.drop('fb_user_9', axis=1)

# As mentioned above, you need to predict churn only for the high-value customers. 
# Define high-value customers as follows: Those who have recharged with an amount more than or equal to X, 
# where X is the 70th percentile of the average recharge amount in the first two months (the good phase).
# => need to consider the columns tot_recharge and avg_rech
# vamsi - av_rech_amt_data => we need to consider for 6, 7 and 8 months (8th month missing value can be replaced by either min/mean -
# need to perform univariate analysis and confirm)
telecom_data = telecom_data.drop('av_rech_amt_data_9', axis=1)
telecom_data = telecom_data.drop('av_rech_amt_data_8', axis=1)

# vamsi - as 8th month is an action phase - missing values can be replaced by 0
# applies to all the below columns values
telecom_data = telecom_data.drop('count_rech_3g_9', axis=1)
telecom_data = telecom_data.drop('count_rech_3g_8', axis=1)

telecom_data = telecom_data.drop('count_rech_2g_9', axis=1)
telecom_data = telecom_data.drop('count_rech_2g_8', axis=1)

telecom_data = telecom_data.drop('max_rech_data_9', axis=1)
telecom_data = telecom_data.drop('max_rech_data_8', axis=1)

telecom_data = telecom_data.drop('total_rech_data_9', axis=1)
telecom_data = telecom_data.drop('total_rech_data_8', axis=1)

telecom_data = telecom_data.drop('date_of_last_rech_data_9', axis=1)
telecom_data = telecom_data.drop('date_of_last_rech_data_8', axis=1)

round(100*(telecom_data.isnull().sum()/len(telecom_data.index)), 2)
```
