```python
import chardet
import pandas as pd

def find_encoding(fname):
    r_file = open(fname, 'rb').read()
    result = chardet.detect(r_file)
    charenc = result['encoding']
    return charenc
```

```python
my_encoding = find_encoding('companies.txt')

print(my_encoding)
```

```python
my_encoding = find_encoding('rounds2.csv')
print(my_encoding)
```
